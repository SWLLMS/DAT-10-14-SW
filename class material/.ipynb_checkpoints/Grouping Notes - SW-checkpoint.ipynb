```python
grouped = sales.groupby('Category')['Sales'].mean().to_frame().rename({'Sales': 'Cat_Avg'}, axis = 1)
### Sim to adding meta data
###.to_frame() turns a series into a dataframe
###.rename to rename the colm you created on axis 1
```

```python
sales.merge(grouped, on='Category', how='left') ###Error if you dont merge to dataframes
```

```python
### Can perform add and subtract from dates

three_mo_offset = pd.DateOffset(months=3)
```

```python
sales.groupby('Customer Name')['Ship Date'].min() + three_mo_offset 
###gives you the 3 month anniversary of when the maide the first purchase
```
