```python
fig, ax = plt.subplots(1, 2)
df[df.Sex == 'female'].plot('Age', 'Fare', kind='Scatter', ax=ax[0], color='red', label='Females')
df[df.Sex == 'male'].plot('Age', 'Fare', kind='Scatter', ax=ax[1], label='Males')

###Must inlcude agument for ax and tell what particular chart axis you want to apear in
```

```python
# you can use the subplots command to set parameters for all charts
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(12, 4))
df[df.Sex == 'female'].plot('Age', 'Fare', kind='Scatter', ax=ax[0], color='red', label='Females')
df[df.Sex == 'male'].plot('Age', 'Fare', kind='Scatter', ax=ax[1], label='Males')
```

```python
###Make a chart and save it as a variable and then pass it in to the second chart see 'ax=ax'
ax = df[df.Sex == 'female'].plot('Age', 'Fare', kind='Scatter', color='red', label='Females')
df[df.Sex == 'male'].plot('Age', 'Fare', kind='Scatter', ax=ax, label='Males')
```

```python
###A much easier way to do more elaborate plotting. Often the best choice for doing exploratory data analysis
#Built off of Matplotlib
#Has the following benefits:
    #Built in methods for using column values to adjust the following visual parameters:
    #Default styling palettes that are tasteful
    #Specialty charts that extend basic matplotlib:
    #heatmap
    #jointplot
    #pairplot
    #regression plots
```

```python
#Primary Seaborn Chart Types:
    #catplot - cateforical data with 9 different viar to add vis diveristy. one must be catagorical and one has to be numeric
    #relplot
    #lmplot/regplot
    #distplot/jointplot
    #pairplot
```

```python
# basic syntax for Seaborn - must specify x and y variables, as well as data source
sns.catplot(x='Pclass', y='Age', data=df)
```

```python
# if you want to switch chart type, just use the 'kind' argument
sns.catplot(x='Pclass', y='Age', kind='swarm', data=df)
```

```python
# if you want to switch chart type, just use the 'kind' argument; box is AKA candle chart
#swarm, strip, box, boxen, violin, point, bar, count
sns.catplot(x='Pclass', y='Age', kind='box', data=df)
```

```python
#Other Useful Arguments in Seaborn: hue, col/row

# let's take a look at the same graph, but using hue to distinguish if they survived
sns.catplot(x='Pclass', y='Age', kind='swarm', hue='Survived', data=df)
```

```python
# you can get a similar effect, but using each value as a separate column
sns.catplot(x='Pclass', y='Age', kind='swarm', hue='Survived', col='Sex', data=df)
```

```python
# or as a row
sns.catplot(x='Pclass', y='Age', kind='swarm', row='Sex', data=df)
```
