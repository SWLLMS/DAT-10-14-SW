```python
### np.select() is for more complex

conditions = [
    (df['Survived'] == 1) & (df['Sex'] == 'female'),###Enter exactly like query for database
    (df['Survived'] == 1) & (df['Sex'] == 'male'),
    (df['Survived'] == 0) & (df['Sex'] == 'female'),
    (df['Survived'] == 0) & (df['Sex'] == 'male'),
]
results = [
    'Female-Survived',
    'Male-Survived',
    'Female-Died',
    'Male-Died'
]

df['Status'] = np.select(conditions, results, 'Other')
```

```python
df['Status'].value_counts()
```

```python
### df.apply use as last resort creates a python function and applys it to dataset

def over_75(val):
        if val > 75:
            return 'Getting up there'
        else:
            return ' You are young'
```

```python
%timeit df['Demographic'] = df['Age'].apply(over_75)
```

```python
def get_status(val): ### val represents every single row in the database
    if val['Survived'] == 1 and val['Sex'] == 'female': ### if val in survived col = 1 and val in sex colm = female
        return 'Female Survived'
    elif val['Survived'] == 0 and val['Sex'] == 'female':
        return 'Female Died'
    elif val['Survived'] == 1 and val['Sex'] == 'male':
        return 'Male Survived'
    elif val['Survived'] == 0 and val['Sex'] == 'male':
        return 'Male Died'
    else: 
        return 'Other'
```

```python
%timeit df['Status'] = df.apply(get_status, axis=1) ### This takes some much longer to do this method.
```
