```python
from sklearn.model_selection import cross_val_score
```

```python
X = df.iloc[:, :-1]
y= df['PRICE']
```

```python
leaf_size = [1, 5, 10, 15, 25]
cross_val_scores =[]

for size in leaf_size:
    tree.set_params(min_samples_leaf=size)
    scores = cross_val_score(estimator = tree, X=X, y=y, cv=10)
    cross_val_scores.append((np.mean(scores), size))
```

Commonly used technique to curb overfitting in your models.
     
     a. Does so by deliberately injecting a measured amount of bias into your model in a way that is mathmatically precise. 
     b. You call bias injection a â€˜penaltyâ€™
     c. Result is that it shrinks your weights
     d. Also known as â€˜weightâ€™ decay
     e. Used in most algorithms that use gradient descent for optimization

#### L1 - Lasso: cost = sum(y - guess)**2 + sum(of absolute value of all weights)
    Penalty term = absolute value of weights
    Effects on Weight: Will weaken and eventually set all to 0 
    Best For: Curbing  Variance and performing feature selection in high dimensional DB
    Typical Use: Overfitting , with dimensaionality close to or greater than length
    * Performs feature selection for you by removing low impact colms. 

#### L2 - Ridge: cost = sum(y - guess)**2 + sum(weights**2)
    Penalty term = Square of weights
    Effect on Weights: will weaken, but not set to 0
    Best For: Curbing variance, but not changing overall composition of variables
    Typical Use: Overfitting, but dataset is taller than iti is wide
    
    
#### Final Piece - Regularization Contstant (Regularized) cost = sum(y - guess)**2 + (alpha * sum(weights**2))

You are using the dirivitive of the weights. The best way to mini alpha is to set weights to 0 or basic linear regression. So only the most impactful values are predicted
Alpha is a number that you set. Has different names:
    a. Alpha
    b. Lambda
    c. C
Determines the strength of your regularization
Common parameter to test in cross validation
Similar to the random forest loop block we made.
