### Step 1).  Load in the `housing.csv` file

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
plt.style.use('seaborn-poster') 
```

```python
df = pd.read_csv('/Users/swllms/DAT-10-14-SW/class material/Unit3/Data/housing.csv')
```

```python
df.head()
```

### Step 2).  Import and initialize the `LinearRegression` algorithm from `SKlearn`

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()
```

### Step 3).  Fit an instance of the `LinearRegression` algorithm using *at least* 1 variable, but with more if you choose.

```python
X = df[['RM', 'CRIM']]
y = df['PRICE']
```

```python
lreg.fit(X,y)
```

### Step 4).  Verify your coefficients and intercept term, and make a column for your predictions

```python
lreg.coef_ #coefficients RM & CRIM
```

```python
lreg.intercept_ #Intercept term
```

```python
df['PREDICTIONS']=lreg.predict(X) #make a column for your predictions
```

```python
df.head()
```

### Step 5).  Score your model

```python
lreg.score(X, y)
```

### Step 7a). Residuals

Create an `Error` column that's the difference between the predictions of your best model and the `PRICE` column.

```python
#Error colm should: 1.) on average valuse of 0 and 2.)Normally Distributed

df['ERROR'] = df['PRICE'] - df['PREDICTIONS']
```
