```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
plt.style.use('seaborn-poster') 
```

```python
df = pd.read_csv('/Users/swllms/DAT-10-14-SW/class material/Unit3/Data/housing.csv')
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg=LinearRegression()
```

```python
X = df.iloc[:, :13]
```

```python
y = df['PRICE']
```

```python
X
```

```python
lreg.fit(X, y)
```

```python
lreg.score(X, y) #74% 
```

```python
lreg.coef_
```

```python
coeffs = pd.DataFrame({
    'Variable': X.columns,
    'Weight' : lreg.coef_
}).sort_values(by='Weight', ascending=False)
```

```python
coeffs #For every value of variable add the value of weight. 
#example for every RM(room) add 3.808865
#A coeff is the value if increased by exactly 1
#the table makes all values weighted the same
#shows the marginal impact of each variable to effect y. 
```

```python
#STANDARDIZING NOTES: Critical step for setting up many models correctly
#Gives all numeric variables a mean of 0, and a variance of 1
#Puts all weights on an equal footing
#Do this by subtracting each value by its mean & divide it by its standard deviation.

#This is the manual way to do it. Use Scalers to preprocess this: See bonus info up tomorrow. 
```

```python
(X - X.mean()).describe() #Called centering the data. 
```

```python
X_std = (X - X.mean()) /X.std() #this is how your data should look before feeding it into model
#hovers around 0 and numberic range is small. 
```

```python
X_std.describe() #std is now 1 across the board, Need to make sure data is all on the same scale
```

```python
lreg.fit(X_std, y)
```

```python
lreg.coef_
```

```python
xcoeffs = pd.DataFrame({
    'Variable': X_std.columns,
    'Weight' : lreg.coef_
}).sort_values(by='Weight', ascending=False)
xcoeffs
```

```python
lreg.score(X_std, y) 
#if your r2 did not change your predictions would be practically the same
#the range is reduce TAX and STAT are now larger than NOX
#this is a better way to read and see the coeffs. 
#Remeber and increase of 1 is an increase of 1 standard d.
#if you dont take this step (Standardizing) your results will not be valid. 

#note this is not to be confused with normalizing, they are different. 
```

```python
#When does standardization apply: 
    #Linear models
    #Anything that uses a weight penalty (l1, l2, etc)
    #Any algorithm that uses gradient descent
```

```python
#Cross Validation Notes: 
#How do we know our model will generalize to the ourside world? 
#We don't know how well our information will work with uknown data. 
#Never fit your model to all of your data. #use cross validation to validate your data.
```

```python
#Bias vs. Variance Tradeoff: looking for the optiomum model
    #Bias refer to when your data doesnt adequately capture the info necessary to solve our problem 
    #Variance refer to when your model mistaknely inerprets random correclations as meaningful 
        #(r2 value looks high but when new data comes in it crashes)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2019)
```

```python
#train and text are used to split the data where the test size is 20% 
#the random state ensure that our random numbers are generated the same 
#(randomly shuffled the same way for everyone)
#nothing special about the yare 2019, you can shuffle by any number
X_train.head()
```

```python
lreg.fit(X_train, y_train) #fit on the training sets
```

```python
lreg.score(X_test, y_test) #score on the test sets
```

```python
#additional step is to create a validation set: test set within the test set. You do not fit on the validation set.
#Think of it as a dress rehersal for the real test set. You would score on the validation set
#remember you may not always have the test set (Because this would come later) 
#and this is where you would use validation set
```
